```python
import itertools
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
```

```python
from scipy.spatial import ConvexHull
points = np.random.rand(30, 2)   # 30 random points in 2-D
hull = ConvexHull(points)
```

```python
plt.plot(points[:,0], points[:,1], 'o')
for simplex in hull.simplices:
    plt.plot(points[simplex, 0], points[simplex, 1], 'k--')
```

```python
from pyhull import convex_hull
```

```python
hull = convex_hull.ConvexHull(points)
```

```python
plt.plot(points[:,0], points[:,1], 'o')
for simplex in hull.simplices:
    plt.plot(simplex.coords[:,0], simplex.coords[:,1], 'k--')
```

```python
from scipy.spatial import Delaunay
```

```python
tri = Delaunay(points)
```

```python
plt.triplot(points[:,0], points[:,1], tri.simplices.copy())
plt.plot(points[:,0], points[:,1], 'o')
plt.show()
```

```python
print tri.find_simplex(points[0]) >= 0
print tri.find_simplex(points[0]-100) >- 0
```

```python
from pyhull import delaunay
```

```python
tri = delaunay.DelaunayTri(points)
```

```python
dim=2
for s in tri.simplices:
    for data in itertools.combinations(s.coords, dim):
        data = np.array(data)
        p, = plt.plot(data[:,0], data[:,1], 'g-')
plt.plot(points[:,0], points[:,1], 'o')
```
